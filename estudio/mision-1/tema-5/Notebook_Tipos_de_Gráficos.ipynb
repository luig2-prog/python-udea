```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Cargar el conjunto de datos Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Mapear las especies a nombres
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Gráfico de barras de la cantidad de cada especie
plt.figure(figsize=(8, 5))
sns.countplot(x='species', data=df)
plt.title('Cantidad de especies')
plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.show()

# Gráfico de caja para ver la distribución del largo del sépalo por especie
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=df)
plt.title('Distribución del largo del sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Largo del sépalo (cm)')
plt.show()

# Gráfico de densidad para el largo del sépalo
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['species'] == 'setosa']['sepal length (cm)'], label='Setosa', fill=True)
sns.kdeplot(df[df['species'] == 'versicolor']['sepal length (cm)'], label='Versicolor', fill=True)
sns.kdeplot(df[df['species'] == 'virginica']['sepal length (cm)'], label='Virginica', fill=True)
plt.title('Densidad del largo del sépalo por especie')
plt.xlabel('Largo del sépalo (cm)')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```
